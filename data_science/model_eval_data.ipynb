```python
import pandas as pd
import numpy as np
```

```python
PATH = '../data/model_metric/'
```

```python
model = pd.read_csv(PATH + 'model_eval.csv')
model
```

```python
rec = model[model.Recommendation == 1].iloc[:,:-3]
rec.dropna(axis=1, inplace=True)
mean_row = rec.iloc[:,1:].apply(np.mean).to_frame().T
mean_row['Model'] = 'Mean'
```

```python
rec_df = pd.concat([rec, mean_row])
rec_df
```

```python
conv = model[model.Conversation == 1].iloc[:,:-3]
conv.dropna(axis=1, inplace=True)
mean_row = conv.iloc[:,1:].apply(np.mean).to_frame().T
mean_row['Model'] = 'Mean'
```

```python
conv_df = pd.concat([conv, mean_row])
conv_df
```

```python
rec_df.columns[1:]
```

```python
def add_percent(series:pd.Series):
    max_num = series.max()
    percent = series/max_num
    percent = percent.map(lambda x:f' ({x:.2%})')
    series = series.astype(str)
    max_len = series.iloc[:-1].apply(len).max()
    series.iloc[-1] = series.iloc[-1][:max_len]
    return series + percent
```

```python
def change_cols(df_:pd.DataFrame):
    df = df_.copy()
    cols = df.columns[1:].tolist()
    for col in cols:
        df[col] = add_percent(df[col])
    return df
```

```python
rec_df = change_cols(rec_df)
conv_df = change_cols(conv_df)
```

```python
rec_df['Recommend'] = 1
conv_df['Conversation'] = 1
```

```python
new_df = pd.concat([rec_df, conv_df])
new_df
```

```python
new_df.to_csv(PATH + 'model_eval_percent.csv', index=False)
```
