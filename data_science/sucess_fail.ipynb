```python
import pandas as pd
# import numpy as np
```

```python
PATH = '../data/durecdial/'
df = pd.read_csv(PATH + 'train_sentence_small.csv')
```

```python
def sucess_fail(df, show_question=True, show_fail = True, min_range=0, max_range=100):
    dfa = df[(df.recdial == 1)][['user_id', 'sentence_index', 'sentence', 'is_user', 'deny']].copy()
    cols = ['user_id', 'sentence_index', 'sentence', 'is_user']
    indices = dfa[(dfa.is_user == 1) & (dfa.deny == show_fail)].index.to_list()
    if show_question:
        new_indices = []
        for index in indices:
            new_indices.append(index-1)
            new_indices.append(index)
        return df.loc[new_indices][cols]
    else:
        return dfa.loc[indices][cols]
```

```python
sucess_fail(df)
```
